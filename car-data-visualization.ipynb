# Visualization of the data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read the previous file ( processed data )
df=pd.read_csv("data/vehicles-phase2.csv")
```

```python
# Convert float to int
df['year']=df['year'].astype('int64')

# Remove id column
df.drop(['id'],axis=1,inplace=True)
df.head()
```

```python
num_col = ['year','odometer']
cat_col = ['region','manufacturer','model','condition','cylinders','fuel','title_status','transmission','drive','size','type','paint_color']
```

```python
# Visualize 100 samples
df2=df.sample(100)

sns.pairplot(df2)
plt.title('Figure 1: Pair Plot')
plt.tight_layout()
plt.savefig('images/visual-pair-plot',dpi=None)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='fuel',y='price',data=df, hue='fuel')
plt.title('Figure 2')
plt.savefig('images/visual-fuel-price',dpi=None)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='fuel',y='price',hue='condition',data=df)
plt.title('Figure 2')
plt.savefig('images/visual-fuel-price-by-condition',dpi=None)
plt.show()
```

```python
fig,axs=plt.subplots(ncols=2)
fig.set_size_inches(12,4)
sns.barplot(x='condition',y='price',data=df,ax=axs[0])
sns.barplot(x='condition',y='price',hue='size',data=df,ax=axs[1])
axs[0].set_title('Figure 5')
axs[1].set_title('Figure 6')
plt.tight_layout()
plt.savefig('images/visual-fuel-price-by-size',dpi=600)
plt.show()
```

```python
fig,axs=plt.subplots(nrows=2)
fig.set_size_inches(15,8)
sns.barplot(x='type',y='price',data=df,ax=axs[0])
sns.barplot(x='manufacturer',y='price',data=df,ax=axs[1])
axs[1].set_xticklabels(df.manufacturer.unique(),rotation=91)
axs[0].set_title('Figure 9')
axs[1].set_title('Figure 10')
plt.tight_layout()
plt.savefig('images/visual-type-manufracture-price',dpi=600)
plt.show()
```
